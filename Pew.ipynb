1) How often do people who have access the internet on a cell phone, tablet or other mobile handheld device use the internet?

Hypothesis : Most of the people that have access to the internet on their phone use it frequently.

Making us of the dplyr functions groupby, mutate and filter to find the count of each of these categories of internet usage.
1   Almost constantly
2   Several times a day
3	About once a day
4	Several times a week, OR
5	Less often



First extract all the users who have internet access and then tabulate the count of how often they use it.

```r
topic1 <- topic %>% select(intmob,intfreq)
topic1 <- filter(topic1, intmob==1)
topic1 <- topic1 %>% group_by(intmob, intfreq) %>% mutate(count = n())
topic1 <- head(unique(topic1),5)
#topic1
```

Plotting the values above to study the distribution.

```r
color <- c("plum","plum2","plum3","plum4")
counts <- topic1$count
barplot(counts, main="Usage Of Internet",beside = TRUE,
  names.arg=c("Constantly", "Less", "Daily Often","Weekly","Once a day"),
        xlab = "FREQUENCY OF USE",ylab = "NUMBER OF USERS",col ="orchid4", border = "black", 
        density=c(100,80,60,40,20), width=c(1,1), xlim=c(0,7))
```

The above bar graph shows that people with internet on their phone do use the internet more.

2) How do people that have a smart phone access internet?

Hypothesis : Most of the people that have smart phones need high speed internet to support the usage of their smart phone.

The question divides people into the following categories and populates them.
1   Dial-up
2	Higher-speed
3	Both Slow-speed/Dial-up and Higher-speed/Broadband

We have to filter the columns smart1 and bbhome1 to arrive at this answer. The count of each category is arrived by grouping the columns and counting the number of appearances. The final column is then mutated to the original dataframe.

```r
topic2 <- topic %>% select(smart1, bbhome1)
topic2 <- filter(topic2, smart1==1)
topic2 <- topic2 %>% group_by(smart1, bbhome1) 
topic2 <- topic2 %>% mutate(count = n())
topic2 <- head(unique(topic2),3)
#topic2
```

Plotting the values above to study the distribution.

```r
color <- c("paleturquoise3","paleturquoise4")
counts <- topic2$count
barplot(counts, main="Smart Phone VS Speed Of Internet",
  names.arg=c("HighSpeed","LowSpeed","Both"),
        xlab = "NUMBER OF USERS",ylab = "SPEED OF INTERNET",col = "indianred",horiz ="true", 
        density=c(70,50,30),width=c(1,1), ylim=c(0,6))
```

From the above graph we can see that, the assumed hypothesis is true. Smart phone users do need high speed internet.

3) How does age affect the use of dating sites among people who have access to the internet?

Hypothesis : As age increases, the usage of dating sites among individuals significantly decreases.

Here I am making use of three columns age, people who have access to the internet and people who visit dating sites to prove my hypothesis.

The age column is first filtered as above 50 years old or below. Then based on their age, the frequency of their use of dating sites is calculated based on dplyr's filter function.

```r
topic3 <- topic %>% select(age,eminuse,date1a)
topic3 <- topic3 %>% filter(age > 50)
topic3 <- topic3 %>% filter(eminuse == 1)
topic3yes <- nrow(topic3 %>% filter(date1a == 1))
topic3no <- nrow(topic3 %>% filter(date1a == 2))
topic3a <- topic %>% select(age,eminuse,date1a)
topic3a <- topic3a %>% filter(age < 50)
topic3a <- topic3a %>% filter(eminuse == 1)
topic3ayes <- nrow(topic3a %>% filter(date1a == 1))
topic3ano <- nrow(topic3a %>% filter(date1a == 2))
YES <- c(topic3yes,topic3ayes)
NO <- c(topic3no,topic3ano)
df <- data.frame(YES,NO)
df<- as.data.frame(df)
```

Plotting the values above to study the distribution

```r
colours <- c("lightblue", "mistyrose")
barplot(as.matrix(df), main="Dating Sites VS Age", ylab = "Number of Users", cex.lab = 1.5, cex.main = 1.4,
        beside=TRUE, col=colours,,density=c(40,40),width=c(1,1), xlim=c(0,10))
legend("topleft", c("Age>50","Age<50"), cex = 1.3, fill = colours)
```

For the above graph we can observe that for age<50 there are more users that use dating sites when compared to age>50.

4) How many people employed full time/part time in a large organisation are full time/part time students?

Hypothesis : People with employment(part time/full time) are less likely to be students.

We have to sift through three columns namely emplnw, emptype1, stud to achieve this. After performing the cleaning we are left with the following values
a) Full time employed - Not students
b) Full time employed - PartTime students
c) Full time employed - FullTime student
d) Part time employed - Not students
e) Part time employed - PartTime students
f) Part time employed - FullTime student


```r
topic4 <- topic %>% select(emplnw, emptype1, stud)
topic4 <- topic4 %>% filter(emplnw == 1 || 2,emptype1 == 1, stud)
a <- nrow(filter(topic4,emplnw==1,stud==1))
b <- nrow(filter(topic4,emplnw==2,stud==1))
c <- nrow(filter(topic4,emplnw==1,stud==2))
d <- nrow(filter(topic4,emplnw==2,stud==2))
e <- nrow(filter(topic4,emplnw==1,stud==3))
f <- nrow(filter(topic4,emplnw==2,stud==3))
val <- c(a,b,c,d,e,f)
```

Plotting the values above to study the distribution

```r
colors <- c("darkslategray","darkslategray","darkslategray")
employ <- c("FullTime Employed","PartTimeEmployed")
student <- c("FullTime Student","PartTimeStudent","Not Student")

# Create the matrix of the values.
Values <- matrix(val,nrow = 3,ncol = 2,byrow = TRUE)
barplot(Values,main = "Employment VS Education",names.arg = employ,xlab = "Employment Type",ylab = "Student Type",
   col = colors,density=c(20,60,100),width=c(1,1), xlim=c(0,4))

# Add the legend to the chart.
legend("topright", student, cex = 1.3, fill = colors,density=c(30,60,100))

```

As we can see from above there are very few people who are employed full time and are also full time students.Majority of the recievers arent students and are full time employed. 
A small percentage of the part time employed lot is a full time/ part time student. 

5) How many people who play video games agree that most people who play video games are men?

Hypothesis : There is a general notion that only men play video games and women cant play them.

To find out the opinion of the play that actually play video games, we have to consider the columns game1 which gives if a person has ever played video games and game3a which gives the general opninion of the public that think mostly men play video games.

```r
topic5 <- topic %>% select(game1, game3a)
topic5 <- topic5 %>% filter(game1 == 1)
total <- nrow(topic5)
agree <- nrow(filter(topic5, game3a == 1))
disagree <- total - agree
opinion<- c(agree,disagree)
```

Plotting the values above to study the distribution

```r
color <- c("paleturquoise3","paleturquoise4")
barplot(opinion, main="Mostly Men Play Video Games",beside = TRUE,
  names.arg=c("Agree","Disagree"),
        xlab = "NUMBER OF USERS",ylab = "OPINION",col = color, 
        density=c(70,50),width=c(1,1), xlim=c(0,7))
```

As we can see from above, most people agree that women dont play as much video games as men.
